```python
def readCoordinatesFromFile(file_name):
    coordinates = []
    min_x = float('inf')
    min_y = float('inf')
    max_x = float('-inf')
    max_y = float('-inf')

    with open(file_name, 'r') as file:
        for line in file:
            point1, point2 = map(str.strip, line.split(" -> "))

            x1, y1 = map(int, point1.split(","))
            x2, y2 = map(int, point2.split(","))

            min_x = min(min_x, x1, x2)
            min_y = min(min_y, y1, y2)
            max_x = max(max_x, x1, x2)
            max_y = max(max_y, y1, y2)

            coordinates.append(((x1, y1), (x2, y2)))

    return coordinates, (min_x, min_y), (max_x, max_y)
```

```python
def createCoordinateMatrix(min_coord, max_coord, default_value='.'):
    min_x, min_y = min_coord
    max_x, max_y = max_coord

    num_rows = max_y - min_y + 1
    num_cols = max_x - min_x + 1

    matrix = [[default_value for _ in range(num_cols)] for _ in range(num_rows)]

    return matrix
```

```python
def incrementCoordinatesOnHorizontalAndVerticalLines(coordinates, matrix, min_coord):
    for coordinateTuple in coordinates:
        point1, point2 = coordinateTuple

        x1, y1 = point1
        x2, y2 = point2

        if y1 == y2:  # Horizontale Linie
            min_x = min(x1, x2)
            max_x = max(x1, x2)
            for x in range(min_x, max_x + 1):
                adjusted_x = x - min_coord[0]
                adjusted_y = y1 - min_coord[1]
                if matrix[adjusted_y][adjusted_x] == '.':
                    matrix[adjusted_y][adjusted_x] = 1
                elif matrix[adjusted_y][adjusted_x] > 0:
                    matrix[adjusted_y][adjusted_x] += 1
        elif x1 == x2:  # Vertikale Linie
            min_y = min(y1, y2)
            max_y = max(y1, y2)
            for y in range(min_y, max_y + 1):
                adjusted_x = x1 - min_coord[0]
                adjusted_y = y - min_coord[1]
                if matrix[adjusted_y][adjusted_x] == '.':
                    matrix[adjusted_y][adjusted_x] = 1
                elif matrix[adjusted_y][adjusted_x] > 0:
                    matrix[adjusted_y][adjusted_x] += 1
```

```python
def count_repeated_coordinates(matrix):
    count = 0
    for row in matrix:
        for value in row:
            if value != '.' and value > 1:
                count += 1
    return count
```

```python
from collections import Counter

file_name = 'input.txt'
coordinates, min_coord, max_coord = readCoordinatesFromFile(file_name)
matrix = createCoordinateMatrix(min_coord, max_coord)
incrementCoordinatesOnHorizontalAndVerticalLines(coordinates, matrix, min_coord)
print(f"Number of different values in the area: {dict(sum(map(Counter, matrix), Counter()))}")
print(f"Summed duplicate points: {count_repeated_coordinates(matrix)}")
print(f"Plotting diagram from {min_coord} to {max_coord}")
print('\n'.join([''.join(['{:1}'.format(item) for item in row]) 
      for row in matrix]))
```
